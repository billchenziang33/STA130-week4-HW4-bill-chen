2.The steps to create a 95% confidence interval.

1.Calculate the sample mean: This is the average value from your sample data.

2.Calculate the SEM: The SEM is obtained by dividing the sample's standard deviation by the square root of the sample size.

3.Determine the margin of error: Multiply the SEM by approximately 2 (the z-value for a 95% confidence interval is roughly 1.96, which is often rounded to 2 for simplicity).

4.Construct the confidence interval: Add and subtract the margin of error from the sample mean to get the lower and upper bounds of the interval.

The result is the interval:
Mean±2×SEM

3.The steps to create 95% confidential interval:

1.Create some random samples from the original sample.

2.caculate the mean of each samples.Then come out the distribution of the sample means.

3.Use the code np.quantile()to find the 2.5% percentile and 97.5% percentile.

4.The range between this two percentile is the answer we are looking for.

````python
import numpy as np

# Simulate some data (e.g., a sample of size n)
np.random.seed(33)  # For reproducibility
data = np.random.normal(loc=100, scale=10, size=30)  # Sample data

# Number of bootstrap samples
n_bootstraps = 1000
bootstrapped_means = []

# Generate bootstrap samples and calculate their means
for _ in range(n_bootstraps):
    bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
    bootstrapped_mean = np.mean(bootstrap_sample)
    bootstrapped_means.append(bootstrapped_mean)

# Convert list to NumPy array for easier calculations
bootstrapped_means = np.array(bootstrapped_means)

# Calculate the 2.5th and 97.5th percentiles
lower_bound = np.quantile(bootstrapped_means, 0.025)
upper_bound = np.quantile(bootstrapped_means, 0.975)

# Print the results
print("Original Sample Mean:", np.mean(data))
print("95% Bootstrapped Confidence Interval:", (lower_bound, upper_bound))
````

````python
import numpy as np

# Create a sample dataset
np.random.seed(33)  # For reproducibility
data = np.random.normal(loc=100, scale=10, size=30)  # give me a sample set of normal distribution.

# Number of bootstrap samples
n_bootstraps = 1000

# Function to calculate bootstrap confidence intervals
def bootstrap_ci(data, n_bootstraps, statistic_func, alpha=0.05):
    bootstrapped_statistics = []

    # Generate bootstrap samples and calculate the specified statistic
    for _ in range(n_bootstraps):
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        stat = statistic_func(bootstrap_sample)
        bootstrapped_statistics.append(stat)

    # Convert list to NumPy array for easier calculations
    bootstrapped_statistics = np.array(bootstrapped_statistics)

    # Calculate the lower and upper percentiles for the confidence interval
    lower_bound = np.quantile(bootstrapped_statistics, alpha/2)
    upper_bound = np.quantile(bootstrapped_statistics, 1 - alpha/2)

    return lower_bound, upper_bound

# Calculate 95% CI for the mean
mean_ci = bootstrap_ci(data, n_bootstraps, np.mean)
print("95% Bootstrap Confidence Interval for the Mean:", mean_ci)

# Calculate 95% CI for the median
median_ci = bootstrap_ci(data, n_bootstraps, np.median)
print("95% Bootstrap Confidence Interval for the Median:", median_ci)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Create the data
data = {
    'InitialHealthScore': [84, 78, 83, 81, 81, 80, 79, 85, 76, 83],
    'FinalHealthScore': [86, 86, 80, 86, 84, 86, 86, 82, 83, 84]
}

# Calculate the differences between Final and Initial scores
diff = np.array(data['FinalHealthScore']) - np.array(data['InitialHealthScore'])

# Bootstrap parameters
n_bootstraps = 1000
np.random.seed(33)

# Function to bootstrap the mean difference
def bootstrap_mean_difference(diff, n_bootstraps=1000):
    boot_means = []
    for _ in range(n_bootstraps):
        bootstrap_sample = np.random.choice(diff, size=len(diff), replace=True)
        boot_means.append(np.mean(bootstrap_sample))
    return np.percentile(boot_means, [2.5, 97.5]), boot_means

# Calculate the 95% bootstrap confidence interval
ci, boot_means = bootstrap_mean_difference(diff)

# Print the confidence interval
print(f"95% Bootstrap Confidence Interval for the Mean Difference: {ci}")

# Visualization of the bootstrap mean differences
plt.hist(boot_means, bins=30, edgecolor='black')
plt.title('Bootstrap Distribution of Mean Differences')
plt.xlabel('Mean Difference')
plt.ylabel('Frequency')
plt.axvline(ci[0], color='red', linestyle='--', label=f"Lower CI: {ci[0]:.2f}")
plt.axvline(ci[1], color='red', linestyle='--', label=f"Upper CI: {ci[1]:.2f}")
plt.legend()
plt.show()
````
